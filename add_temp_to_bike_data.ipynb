```python
#imports
import pandas as pd
from datetime import datetime
```

```python
# load the data
data_path = 'Datasets/2021-2023/CSB_2021-2023_All.csv'
initial_data = pd.read_csv(data_path)
```

```python
# copy data (it's faster to copy than to load it again)
bike_data = initial_data.copy()
```

```python
# load the data
data_path = 'Datasets/2021-2023/CSB_2021-2023_Hourly_Weather.csv'
weather_data = pd.read_csv(data_path)
```

```python
# convert datetime data
bike_data['started_at'] = pd.to_datetime(bike_data['started_at'])
bike_data['ended_at'] = pd.to_datetime(bike_data['ended_at'])
```

```python
# Concatenate 'Date' and 'Hour' columns and convert to datetime
weather_data['Datetime'] = pd.to_datetime(weather_data['Date'] + ' ' + weather_data['Hour'].astype(str), format='%Y-%m-%d %H')
```

```python
# Round 'started_at' to the nearest hour
bike_data['rounded_started_at'] = bike_data['started_at'].dt.round('H')

# Merge based on the rounded 'started_at' and 'Datetime' columns
merged_data = pd.merge(bike_data, weather_data[['Datetime', 'temperature_2m (°C)', 'relativehumidity_2m (%)']], left_on='rounded_started_at', right_on='Datetime', how='left')

# Drop the redundant columns if needed
merged_data = merged_data.drop(['Datetime', 'rounded_started_at'], axis=1)
merged_data
```

```python
# Create a datetime object for comparison
comparison_datetime = datetime(2022, 1, 1, 22, 0, 0)

# Boolean indexing to filter rows where 'Datetime' is equal to the comparison datetime
filtered_data = weather_data[weather_data['Datetime'] == comparison_datetime]
filtered_data
```
